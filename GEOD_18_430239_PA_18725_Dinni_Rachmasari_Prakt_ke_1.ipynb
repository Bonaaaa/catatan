 1. x(t) = cos(t) + sin (t + $ \frac {\pi}{2}$)  $\;\;\;\;\;\;$  Interval : 0 $\leq$ t $\leq$ 2$\pi$
 2. x(t) = $\frac {1}{\sqrt{2\pi}} e^{- \frac {t^{2}}{0.4}}$  $\;\;\;\;\;\;\;\;\;\;\;$  Interval : -5 $\leq$ t $\leq$ 5
 3. x(t) = $e^{-t}H(t+2)$ $\;\;\;\;\;\;$ Interval : -5$\leq$ t $\leq$ 5
 4. x(t) = $\delta (t-2)$ $\;\;\;\;\;\;$ Interval : -5$\leq$ t $\leq$ 5
 5. x(t) = $\delta (t-1) + \delta (t+2)$ $\;\;\;\;\;\;\;\;\;\;$ Interval : -5$\leq$ t $\leq$ 5

```python
# Jawaban No.1
import numpy as np
import pylab as plt

# Fase sinyal
theta = np.pi / 2
# Periode sinyal
T = 2 * np.pi
# interval waktu
dt = 0.01
t = np.arange(0, T + dt, dt)
# Fungsi sinyal
x = np.cos( t ) + np.sin( t + theta)
# Plot sinyal
plt.plot(t, x)
plt.xlabel('t')
plt.ylabel('x(t)')
plt.show()
```

```python
# Jawaban No.2
import numpy as np
import pylab as plt
import math as mt

# waktu sinyal
dt = 0.1
t = np.arange(-5, 5 + dt, dt)
# Sinyal x(t)
x = (1 / mt.sqrt(2*np.pi)) * np.exp(-t**2 / 0.4)
# Plot sinyal
plt.plot(t, x, 'k', label='x(t)')
plt.show()
```

```python
# Jawaban No.3
import numpy as np
import pylab as plt

# interval waktu
dt = 0.1
t = np.arange(-5, 10 + dt, dt)
# Sinyal x(t)
x = (np.exp(-t)*np.heaviside(t+2,1))
# Plot sinyal
plt.plot(t, x, 'k', label='x(t)')
plt.show()
```

```python
# Jawaban No.4
import numpy as np
import pylab as plt

# definisikan interval waktu pertama
dt = 0.1
t1 = np.arange(-5,-0.1+dt,dt)
# definisikan interval waktu kedua
t2 = [0.]
# definisikan interval waktu ketiga
t3 = np.arange(0.1,10+dt,dt)
# definisikan d(t) pada interval t1
d1 = np.zeros_like(t1)
# definisikan d(t) pada t2
d2 = [1.]
# definisikan d(t) pada interval t3
d3 = np.zeros_like(t3)
# rangkaian vector t dan d
t = np.concatenate((t1,t2,t3))
d = np.concatenate((d1,d2,d3))
#plot sinyal
plt.plot(t+2,d)
plt.show()
```

```python
# Jawaban No.5
import numpy as np
import pylab as plt

# definisikan interval waktu pertama
dt = 0.1
t1 = np.arange(-5,-2.1+dt,dt)
# definisikan interval waktu kedua
t2 = [-2.]
# definisikan interval waktu ketiga
t3 = np.arange(-1.9,0.9+dt,dt)
#definisikan interval waktu ke empat
t4 = [1.]
#definisikan interval waktu ke lima
t5 =  np.arange(1.1,10+dt,dt)
# definisikan  d(t) pada interval t1
d1 = np.zeros_like(t1)
# definisikan  d(t) pada t2
d2 = [1.]
# definisikan  d(t) pada interval t3
d3 = np.zeros_like(t3)
# definisikan d(t) pada t4
d4 = [1.]
# definisikan  d(t) pada interval t5
d5 = np.zeros_like(t5)
# rangkaian vektor t dan d
t = np.concatenate((t1,t2,t3,t4,t5))
d = np.concatenate((d1,d2,d3,d4,d5))
#plot sinyal
plt.plot(t,d)
plt.show()
```
